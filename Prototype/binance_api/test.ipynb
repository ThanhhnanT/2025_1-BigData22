```python
from kafka.admin import KafkaAdminClient, NewTopic

admin_client = KafkaAdminClient(bootstrap_servers="localhost:9092")

topic_name = "orderbook_updatesss"
if topic_name not in admin_client.list_topics():
    # Tạo topic với 20 partitions ngay và luôn
    new_topic = NewTopic(name=topic_name, num_partitions=20, replication_factor=1)
    admin_client.create_topics(new_topics=[new_topic], validate_only=False)
    print(f"✅ Đã tạo topic {topic_name} với 20 partitions")
```
