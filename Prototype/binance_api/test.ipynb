```python
from kafka import KafkaConsumer
import json

consumer = KafkaConsumer(
    'orderbook_update',
    bootstrap_servers=['localhost:9092'],
    auto_offset_reset='earliest',
    value_deserializer=lambda m: json.loads(m.decode('utf-8'))
)

print("üîç ƒêang soi c√°c partition...")

for message in consumer:
    symbol = message.key.decode('utf-8') if message.key else "NO KEY"
    print(f"üì¶ Coin: {symbol} | Partition: {message.partition} | Offset: {message.offset}")
```
